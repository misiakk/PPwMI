```python
from sklearn.model_selection import train_test_split
```

```python
# utworzenie zmiennej X zawierającej zmienne objaśniające oraz zmiennej y jako zmiennej objaśnianej
X = df_insurance[['age', 'bmi', 'children', 'sex_encoded', 'smoker_encoded',
       'region_northwest', 'region_southeast', 'region_southwest']]
y = df_insurance['charges']
```

```python
# utworzenie zbioru treningowego ( do nauki modelu) oraz zbioru testowego (do oceny wydajności modelu) -> test_size = 0.4 -> 40% danych jest używanych do testowania
# natomiast 60% danych jest przeznaczane do treningu modelu (domyślnie dopełnienie do 100% - test_size, jeżelei nie jest ustawiony parametr train_size)
# random_state = 42 -> ustawienie jakiejkolwiek liczby, dzięki czemu zapewnia isę podział deterministyczny i reprodukowalność wyników
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# utworzenie obiektu modelu regresji
lm = LinearRegression()
```

```python
# dopasowanie modelu do danych treningowych
lm.fit(X_train,y_train)
```

```python
# wyprintowanie stałej w modelu 
print(lm.intercept_)
```

```python
# utworzenie ramki danych w celu przedstawienia współczynników modelu regresji liniowej wraz z nazwami zmiennych objaśniających
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
# przeprowaadzenie prognoz na danych testowych przy uzyciu wytrenowanego modelu regresji liniowej, aby ocenić wydajność modelu
predictions = lm.predict(X_test)
```

```python
# utworzenie wykresu punktowego, który pokazuje dopasowanie wartości rzeczywistych z przewidywanymi, oś pozioma -> rzeczywiste, oś pionowa -> predykcje
# mozna dostrzec ze predykcje nie są idealnie dopasowane do wartości rzeczywistych, ponieważ poszczególne obserwacje odbiegają od linii y = x
plt.scatter(y_test,predictions)
```

```python
# pokazanie na wykresie rozkładu różnic między wartościami rzeczywistymi a predykcjami, wykres mozna powiedzieć, 
# że jest zbliżony częściowo do normalnego, co mówi o tym ze model liniowy calkiem dobrze radzi sobie z danymi
sns.distplot((y_test-predictions),bins=50);
```

```python
from sklearn import metrics
```

```python
# obliczenie średniego błędu bezwzględnego (srednie różnice między wartosciami rzeczywistymi i predykcjami) oraz pierwiastek ze sredniego kwadratu błędu  
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
import statsmodels.api as sm
```

```python
# inicjalizacja modelu regresji liniowej przy uzyciu biblioteki statsmodels
model = sm.OLS(y_train, X_train)
```

```python
# dostosowanie modelu do danych treningowych 
results = model.fit()
```

```python
# wyprintowanie podsumowania modelu
# WNIOSKI :
# R^2 i skorygowany R^2 wskazują na wysokie dopasowanie modelu do danych, a wlasciwie mówi, ze 87% zmienności w zmiennej zależnej jest wyjaśniony przez model
# Prop(F-statistics) mówi o odrzuceniu hipotezy zerowej o łącznej nieistotności zmiennych objaśniających 
# test Durbina-Watsona wskazuje na niewystepowanie autokorelacji reszt -> wartosc bliska 2, nie ma znaczacego wplywu na oszacowania, ponieważ nie występuje szereg czasowy
# test Jarque-Bera mówi o tym, czy reszty maja rozklad zblizonhy do normalnego, w tym przypadku dosyc wysoka statystyka testowa, moze swiadczyc o tym ze reszty nie pochodza z rozkladu normanego -> brak konsekwencji, ponieważ N stosunkowo duze
# coef age = 205.9 -> wzrost wieku o 1 rok powoduje wzrost skladki o 205.9 jednostek pieniężnych
# coef bmi = 65.11 -> wzrost bmi o 1 powoduje wzrost skladki o 65.11 jednostek pienięznych
# coef children -> zmienna nieistotna statystycznie, ponieważ p-value > 0.05
# coef sex_encoded = -970.83 -> kobiety płacą o 970.83 jednostek pieniężnych mniejszą składkę w porównaniu do mężczyzn
# coef smoker_encoded = 22740 -> osoby palące w porównaniu do osób niepalących płacą o 22740 jednostek pienięznych większą składkę 
# coef region_northwest = -1432.88 -> osoby zamieszkujące region northwest płacą o 1432.88 mniejszą składkę niż osoby zamieszkujące region northeast
# coef region_southeast = zmienna nieistotna statystycznie, bo p-value>0.05
# coef region_southwest = -1427.34 -> osoby zamieszkujące region southwest płacą o 1427.34 mniejszą składkę niż osoby zamieszkujące region northeast
print(results.summary())
```
